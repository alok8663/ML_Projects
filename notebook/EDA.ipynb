### Data Visualization

```python
# Creating a pie chart
values = df.Exited.value_counts()
labels = ['Not Exited', 'Exited']

fig, ax = plt.subplots(figsize = (4, 3), dpi = 100)
explode = (0, 0.09)

patches, texts, autotexts = ax.pie(values, labels = labels, autopct = '%1.2f%%', shadow = True,
                                   startangle = 90, explode = explode)

plt.setp(texts, color = 'grey')
plt.setp(autotexts, size = 8, color = 'white')
autotexts[1].set_color('black')
plt.show()
```

### Insights


* 20% of customers have churned, indicating that one in every five has stopped using the service or left the company.

```python
# Visualizing Categorical Variables

fig, ax = plt.subplots(3, 2, figsize=(18, 15))

sns.countplot(x='Geography', hue='Exited', data=df, ax=ax[0][0])
sns.countplot(x='Gender', hue='Exited', data=df, ax=ax[0][1])
sns.countplot(x='Tenure', hue='Exited', data=df, ax=ax[1][0])
sns.countplot(x='NumOfProducts', hue='Exited', data=df, ax=ax[1][1])
sns.countplot(x='HasCrCard', hue='Exited', data=df, ax=ax[2][0])
sns.countplot(x='IsActiveMember', hue='Exited', data=df, ax=ax[2][1])

plt.tight_layout()
plt.show()
```

#### Insights

* The majority of customers are from France, but most of those who churned are from Germany.
* The proportion of female customers who churned is higher than that of male customers.
* Majority of customers have a tenure between 1 and 9 years, and the churn rate is also higher within this tenure range.
* Most customers have 1 or 2 products, and the majority of those who churned have only 1 product. This may suggest that they are not satisfied, leading to their churn.
* Majority of customers who churned have credit cards, but this may be coincidental, as most customers have credit cards.
* Inactive members have a higher churn rate, and the overall proportion of inactive members is also quite high.

```python
# Visualizing Continuous Variables

fig, ax = plt.subplots(2, 2, figsize = (16, 10))

sns.boxplot(x = 'Exited', y = 'CreditScore', data = df, ax = ax[0][0])
sns.boxplot(x = 'Exited', y = 'Age', data = df, ax = ax[0][1])
sns.boxplot(x = 'Exited', y = 'Balance', data = df, ax = ax[1][0])
sns.boxplot(x = 'Exited', y = 'EstimatedSalary', data = df, ax = ax[1][1])

plt.tight_layout()
plt.show()
```

### Insights

* There is no significant difference in the credit score distribution between customers who have churned and those who have not.
* Older customers are churning at a higher rate than younger ones.
* The bank is losing customers with significant bank balances.
* Estimated salary does not have a significant impact on the likelihood of churning.

```python
# Heatmap 
df_numeric = df.select_dtypes(include=[np.number])

corr = df_numeric.corr()


plt.figure(figsize = (20, 12))
sns.heatmap(corr, linewidths = 1, annot = True, fmt = ".2f")
plt.show()
```

#### We can conclude that there is no multicollinearity in the data.
